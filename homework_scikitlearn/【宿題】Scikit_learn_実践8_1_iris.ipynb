```python
# データの読み込み
from sklearn.datasets import load_iris
iris = load_iris()
X, y = iris.data, iris.target
```

```python
# 基礎集計からこのデータの説明をしてください。
import pandas as pd
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df.head()
```

```python
# アヤメの種別　【質問】このデータに「種別」が含まれていることをどうやって知るのでしょうか。
iris.target
```

```python
# 種別をデータセットに追加
df['category'] = iris.target
df.head()
```

```python
# データセット数の確認 (5要素の情報を持ったアヤメが150レコード存在する)
df.shape
```

```python
# 基礎集計後、train, testに分割しtestデータを推論してください。
X, y = iris.data, iris.target

from sklearn.model_selection import train_test_split
train_test_split(X, y, test_size=0.3, random_state=5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)
X.shape, X_train.shape, X_test.shape, y.shape, y_train.shape, y_test.shape
```

```python
# モデルはDecisionTreeClassifierと課題と同じランダムフォレストの２種類を使ってください。
# 決定木の予測モデル
from sklearn.tree import DecisionTreeClassifier

# クラスをインスタンス化
model_decisiontree = DecisionTreeClassifier(random_state=3)

# 予測モデルの学習(fitメソッドの引数にデータを指定する)
model_decisiontree.fit(X_train, y_train)
```

```python
# 予測モデルの評価：accuracy(正解率)を算出してください

y_pred_decisiontree = model_decisiontree.predict(X_test)
y_test

from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred_decisiontree)
```

```python
# scoreメソッドで正解率の検証 (scoreメソッドは分類予測の場合は正解率を出力、回帰予測モデルの場合は決定係数を出力)
model_decisiontree.score(X_test, y_test)
```

```python
# ランダムフォレストの予測モデル
from sklearn.ensemble import RandomForestClassifier

# クラスをインスタンス化
model_randomforest = RandomForestClassifier(random_state=3)

# 予測モデルの学習
model_randomforest.fit(X_train, y_train)
```

```python
# 予測モデルの評価：accuracy(正解率)を算出してください
y_pred_randomforest = model_randomforest.predict(X_test)

accuracy_score(y_test, y_pred_decisiontree)
```

```python
# scoreメソッドで正解率の検証
model_randomforest.score(X_test, y_test)
```
