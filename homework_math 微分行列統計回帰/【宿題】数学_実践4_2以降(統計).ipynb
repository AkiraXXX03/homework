```python
# 実践4-3(応用):
# 二項分布・ポアソン分布・指数分布・ベルヌーイ分布・幾何分布について調べて報告してください。

pmf: 離散確率質量関数
pdf: 連続確率密度関数
λ : 単位時間当たりの平均発生回数
p  : 成功確率

from scipy import stats

①ベルヌーイ分布
【特徴】1回のみの試行で「成功（1）」または「失敗（0）」のどちらかが出る確率
【数学】P(X=x) = p^x(1-p)^(1-x) (x=0,1)
【python】stats.bernoulli.pmf(x,p)
          例：成功確率 p=0.3 のとき x=1 の確率
【利用例】広告のクリック確率、True/Falseの2択の確率

②二項分布
【特徴】n回数の試行の中で「成功」が起こる回数を表す（成功確率は一定）
【数学】P(X=k) = (n k)p^k(1-p)^(n-k)
【python】stats.binom.fmp(k,n,p)
         例：n=10 回試行して成功確率 p=0.3、成功回数 k=4 の確率
【利用例】100人に広告を見せてクリックする人数の確率、製品1000個中の不良品確率

③幾何分布
【特徴】初めて成功するまでに何回試行が必要か（成功までの回数）
【数学】P(X=k) = (1-p)^(k-1)*p
【python】stats.geom.pmf(k,p)
         例：成功確率 p=0.2、初成功が 5 回目の確率
【利用例】広告が表示されるまでのクリック回数、顧客が購入するまでの接触回数

④ポアソン分布
【特徴】一定時間・空間内で、ある事象が「何回」発生するか（まれなイベントの回数）
【数学】P(X=k) = λ^k *e^-k / k!
【python】stats.poisson.pmf(k,λ)
         例：λ=2 のとき k=3 の確率
【利用例】地震や事故の発生予測、一定時間内のアクセス数やクレーム数の予測

⑤ 指数分布：
【特徴】連続時間において、次の事象が発生するまでの「待ち時間」
【数学】f(x)=λe ^-λx（x ≥ 0）
【python】stats.expon.pdf(k, p)
         例：λ=2 のとき x=1.0 の確率密度
【利用例】システムの故障間隔、店舗の行列待ち時間、リピーター客の次回購入までの時間モデル
```
