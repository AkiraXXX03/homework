```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
import matplotlib.pyplot as plt

# データセットの読み込み
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
X.head()
```

```python
X.describe()
```

```python
# 階層的クラスタリングの実行(ウォード法)
linked = linkage(X, 'ward')
```

```python
# デンドログラムの表示
plt.figure(figsize=(10, 5))
dendrogram(linked,
           orientation='top',
           labels=iris.target_names[iris.target],
           distance_sort='descending',
           show_leaf_counts=True)
plt.title('Dendrogram of Iris data-set')
plt.show()
```

```python
# クラスタリング結果の取得(４クラスターに分割)
clusters = fcluster(linked, 4, criterion='maxclust')

# 結果表示
for i, cluster in enumerate(clusters):
  print(f"データポイント {i+1}: クラスタ{cluster}")

# クラスタリング結果と元のラベルの比較
comparison_df = pd.DataFrame({'クラスタ': clusters, '品種': iris.target})
print("\nクラスタリング結果と品種の比較:")
print(comparison_df)

# 予測結果と正解データの比較
print("\nクラスタリング結果と品種の比較（品種別カウント）:")
print(comparison_df.groupby('クラスタ')['品種'].value_counts())
```
